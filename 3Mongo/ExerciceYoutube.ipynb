## Question 0

### Netoyer les données

```python
df_youtube['tags'] = df_youtube['tags'].apply(lambda x: x.split("|"))
df_youtube = df_youtube.rename(columns={"video_id" : "_id"})
df_youtube.drop_duplicates(subset ="_id", keep = 'first', inplace=True)
```

```python
df_youtube = df_youtube.drop_duplicates(subset ="_id", keep = 'first', inplace=True)
```

## Question 5

```python
moyenne = collection.aggregate([{"$group" : {"_id" : "$category_id", "numberviewscategoryid" : {"$avg" : "$views"}}}])
list(moyenne)
```

## Question 6 

```python
most_like = collection.aggregate([{"$group" : {"_id" : "$channel_title", "numberlikes" : {"$avg" : "$likes"}}}, 
                                  {"$sort":{"numberlikes":-1}}, 
                                  {"$limit" : 5}
                                 ])
list(most_like)
```
